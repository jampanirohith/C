## Problem 1: Pointers and File Handling

- Traverse the array of numbers in the file, and calculate the sum.
- Print the sum to the console and write the result to `output.txt`.


## Problem 2: Numerical Differentiation
Write a C program to compute the derivative of a function using the backward difference formula.

$$ f'(x) = \frac{f(x) - f(x-h)}{h} $$

**Requirements:**
- Use the function: `f(x) = x^3 - 4x + 1`.
- Take `x` and `h` (step size) as input.
- Compute the derivative using backward difference formula.
- Compare the results with the actual derivative `f'(x) = 3x^2 - 4`.
